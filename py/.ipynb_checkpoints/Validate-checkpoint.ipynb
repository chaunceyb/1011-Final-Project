```python
import pickle as pkl
import pandas as pd
import matplotlib.pyplot as plt
class model_state():
    def __init__(self):
        self.epochs = 0
        self.log = []

    def update(self, train_loss, val_bleu):
        self.epochs += 1
        log = {'epoch': self.epochs, 'train_loss': train_loss, 'val_bleu': val_bleu}
        self.log.append(log)
        return

    def next_epoch(self):
        return self.epochs + 1

    def get_log(self):
        return self.log
```

```python
plt.rc('text', usetex=True)
plt.rc('font', family=**{'family': 'serif', 'serif': ['Times']})
```

```python
log = pkl.load(open('../state/vi_attn_lstm/log.pkl', 'rb'))
```

```python
log_df = pd.DataFrame(log.get_log())
```

```python
%matplotlib inline 
plt.plot(log_df.epoch, log_df.val_bleu, 'k-', label = 'val BLEU', linewidth = 1)
plt.grid(linestyle = '--')
plt.xlabel('Epochs')
plt.ylabel(r'Validation \textsc{BLEU}')
plt.show()
```
